# Hypothesis Testing

In this section, statistical tests are conducted to evaluate whether the relationships observed in the exploratory data analysis are statistically significant.

The goal is to move beyond visual inspection and determine whether observed patterns are likely to be meaningful or due to random variation.

```python
import pandas as pd
import numpy as np
from scipy import stats

merged = pd.read_csv("merged_dataset.csv")
merged.head()
```

## Test 1: Correlation between Inflation and Obesity

H0: There is no relationship between inflation and obesity.

H1: There is a relationship between inflation and obesity.

```python
corr, p_value = stats.pearsonr(
    merged["inflation_rate"],
    merged["obesity_rate"]
)

print("Correlation:", corr)
print("P-value:", p_value)
```

If the p-value is less than 0.05, we reject the null hypothesis and conclude that there is a statistically significant relationship.

Otherwise, we fail to reject the null hypothesis.

The p-value is greater than 0.05, indicating that the relationship between inflation and obesity is not statistically significant. Therefore, we fail to reject the null hypothesis.

## Test 2: Correlation between Fast Food Density and Obesity

H0: There is no relationship between fast food density and obesity.

H1: There is a relationship between fast food density and obesity.

```python
corr2, p_value2 = stats.pearsonr(
    merged["fast_food_per_million"],
    merged["obesity_rate"]
)

print("Correlation:", corr2)
print("P-value:", p_value2)
```

The p-value is greater than 0.05, suggesting that the relationship between fast food density and obesity is not statistically significant. Therefore, we fail to reject the null hypothesis.

```python
median_value = merged["fast_food_per_million"].median()

high_fastfood = merged[merged["fast_food_per_million"] >= median_value]
low_fastfood = merged[merged["fast_food_per_million"] < median_value]
```

```python
t_stat, p_val = stats.ttest_ind(
    high_fastfood["obesity_rate"],
    low_fastfood["obesity_rate"],
    equal_var=False
)

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

## Test 3: Difference in Obesity between High and Low Fast Food Countries

H0: There is no difference in obesity rates between countries with high and low fast food density.

H1: There is a difference.

The p-value is greater than 0.05, indicating that there is no statistically significant difference in obesity rates between countries with high and low fast food density. Therefore, we fail to reject the null hypothesis.

```python
import statsmodels.api as sm

X = merged[["inflation_rate", "fast_food_per_million"]]
y = merged["obesity_rate"]

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

The regression results show that neither inflation nor fast food density is a statistically significant predictor of obesity at the 5% significance level.

This suggests that other factors not included in the model may be more important in explaining obesity differences across countries.

## Final Conclusion

The hypothesis testing results indicate that:

- There is no statistically significant relationship between inflation and obesity.
- There is no statistically significant relationship between fast food density and obesity.
- There is no significant difference in obesity between high and low fast food countries.

These findings suggest that the variables analyzed in this study are not sufficient to explain obesity differences across countries on their own.

Further research including additional variables would be necessary to better understand these relationships.

These results do not imply that no relationship exists, but rather that no statistically significant relationship was detected in this dataset.
