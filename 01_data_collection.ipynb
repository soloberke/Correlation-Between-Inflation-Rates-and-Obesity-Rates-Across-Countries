```
import pandas as pd

inflation = pd.read_csv("inflation.csv", skiprows=4)
obesity = pd.read_csv("obesity.csv")  # buna gerek yok
population = pd.read_csv("population.csv", skiprows=4)
fastfood = pd.read_csv("fastfood.csv")
```

```
inflation_long = inflation.melt(
    id_vars=["Country Name"],
    value_vars=[str(year) for year in range(1960, 2026)],
    var_name="year",
    value_name="inflation_rate"
)
```

```
inflation_long = inflation_long.rename(columns={"Country Name": "country"})
```

```
inflation_long["year"] = pd.to_numeric(inflation_long["year"], errors="coerce")
```

```
inflation_long["inflation_rate"] = pd.to_numeric(inflation_long["inflation_rate"], errors="coerce")
```

```
inflation_long = inflation_long.dropna(subset=["inflation_rate"])
```

```
obesity.head()
obesity.columns.tolist()
```

```
obesity_clean = obesity.rename(columns={
    "Entity": "country",
    "Year": "year",
    "Prevalence of obesity among adults, BMI >= 30 (age-standardized estimate) (%) - Sex: both sexes - Age group: 18+  years of age": "obesity_rate"
})
```

```
obesity_clean = obesity_clean[["country", "year", "obesity_rate"]]
```

```
obesity_clean = obesity_clean[["country", "year", "obesity_rate"]]
```

```
obesity_clean = obesity_clean.dropna(subset=["obesity_rate"])
```

```
population = population.drop(columns=["Unnamed: 70"], errors="ignore")

population_long = population.melt(
    id_vars=["Country Name"],
    value_vars=[str(year) for year in range(1960, 2026)],
    var_name="year",
    value_name="population"
)

population_long = population_long.rename(columns={"Country Name": "country"})

population_long["year"] = pd.to_numeric(population_long["year"], errors="coerce")
population_long["population"] = pd.to_numeric(population_long["population"], errors="coerce")

population_long = population_long.dropna(subset=["population"])

print(population_long.head())
print(population_long.shape)
print(population_long.columns.tolist())
```

```
fastfood_count = (
    fastfood.groupby("country")
    .size()
    .reset_index(name="fast_food_count")
)

print(fastfood_count.head())
print(fastfood_count.shape)
print(fastfood_count.columns.tolist())
```

```
fastfood_count["country"] = fastfood_count["country"].str.strip()
```

```
inflation_long["country"] = inflation_long["country"].str.strip()
obesity_clean["country"] = obesity_clean["country"].str.strip()
population_long["country"] = population_long["country"].str.strip()
fastfood_count["country"] = fastfood_count["country"].str.strip()
```

```
inflation_2022 = inflation_long[inflation_long["year"] == 2022].copy()
obesity_2022 = obesity_clean[obesity_clean["year"] == 2022].copy()
population_2022 = population_long[population_long["year"] == 2022].copy()

print(inflation_2022.shape)
print(obesity_2022.shape)
print(population_2022.shape)
```

```
merged = pd.merge(
    inflation_2022,
    obesity_2022,
    on=["country", "year"],
    how="inner"
)

print(merged.head())
print(merged.shape)
```

```
merged = pd.merge(
    merged,
    population_2022,
    on=["country", "year"],
    how="inner"
)

print(merged.head())
print(merged.shape)
```

```
merged = pd.merge(
    merged,
    fastfood_count,
    on="country",
    how="inner"
)

print(merged.head())
print(merged.shape)
print(merged.columns.tolist())
```

```
merged["fast_food_per_million"] = (
    merged["fast_food_count"] / merged["population"] * 1_000_000
)

print(merged[["country", "fast_food_count", "population", "fast_food_per_million"]].head())
```

```
merged = merged.dropna(subset=[
    "inflation_rate",
    "obesity_rate",
    "population",
    "fast_food_count",
    "fast_food_per_million"
])

print(merged.shape)
```

```
merged.to_csv("merged_dataset.csv", index=False)
```

```

fastfood_count = (
    fastfood.groupby("country")
    .size()
    .reset_index(name="fast_food_count")
)

inflation_long["country"] = inflation_long["country"].str.strip()
obesity_clean["country"] = obesity_clean["country"].str.strip()
population_long["country"] = population_long["country"].str.strip()
fastfood_count["country"] = fastfood_count["country"].str.strip()

inflation_2022 = inflation_long[inflation_long["year"] == 2022].copy()
obesity_2022 = obesity_clean[obesity_clean["year"] == 2022].copy()
population_2022 = population_long[population_long["year"] == 2022].copy()

merged = pd.merge(
    inflation_2022,
    obesity_2022,
    on=["country", "year"],
    how="inner"
)


merged = pd.merge(
    merged,
    population_2022,
    on=["country", "year"],
    how="inner"
)


merged = pd.merge(
    merged,
    fastfood_count,
    on="country",
    how="inner"
)


merged["fast_food_per_million"] = (
    merged["fast_food_count"] / merged["population"] * 1_000_000
)


merged = merged.dropna(subset=[
    "inflation_rate",
    "obesity_rate",
    "population",
    "fast_food_count",
    "fast_food_per_million"
])


merged.to_csv("merged_dataset.csv", index=False)

print(merged.head())
print(merged.shape)
print(merged.columns.tolist())
```
